# Plots Using **Python**

## Load Data

```python
# Load libraries
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# Load data
popDF = pd.read_excel('world-population.xlsm')
```

## Line Chart

```python
# Set up Line chart
popDF.plot(kind='line', x='Year', y='Population', figsize=(10, 6),
          xlabel='Year', ylabel='Population (billions)')

# Set up chart title
plt.title("World Population by Year", fontsize = 15)
```

<div style="page-break-after: always;"></div>

## Step Chart

```python
# Set up step chart
popDF.plot(kind='line', drawstyle='steps-post', x='Year', y='Population', figsize=(10, 6),
          xlabel='Year', ylabel='Population (billions)')

# Set up chart title
plt.title("World Population by Year", fontsize = 15)
```
