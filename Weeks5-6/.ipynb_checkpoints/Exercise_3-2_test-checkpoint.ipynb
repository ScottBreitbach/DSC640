## Tree Map

```python
# Build a treemap w/Seaborn and Squarify
import squarify

sns.set_style(style="whitegrid") # set seaborn plot style
sizes= expenDF["expenditure"].values# proportions of the categories
label=expenDF["category"]
squarify.plot(sizes=sizes, label=label, alpha=0.6).set(title='Treemap with Squarify')
plt.axis('off')
plt.show()
```

```python
size = expenDF.groupby('category')[['expenditure']].sum().expenditure.tolist()
labels = expenDF.groupby('category')[['expenditure']].sum().index.tolist()

cmap = matplotlib.cm.Spectral #.Blues
mini = min(size)
maxi = max(size)
norm = matplotlib.colors.Normalize(vmin=mini, vmax=maxi)
colors = [cmap(norm(value)) for value in size]

squarify.plot(sizes=size, label=labels, alpha=.9, color=colors)
plt.axis('off')
plt.show()
```

#### Control the color of Treemap

```python
#libraries
import matplotlib.pyplot as plt
import squarify # pip install squarify (algorithm for treemap)
 
# Change color
squarify.plot(sizes=[13,22,35,5], label=["group A", "group B", "group C", "group D"], color=["red","green","blue", "grey"], alpha=.4 )
plt.axis('off')
plt.show()
```

#### Treemap with colors mapped on values

```python
#libraries
import matplotlib
import matplotlib.pyplot as plt
import squarify # pip install squarify (algorithm for treemap)&lt;/pre&gt;
 
# Create a dataset:
my_values=[i**3 for i in range(1,100)]
 
# create a color palette, mapped to these values
cmap = matplotlib.cm.Blues
mini=min(my_values)
maxi=max(my_values)
norm = matplotlib.colors.Normalize(vmin=mini, vmax=maxi)
colors = [cmap(norm(value)) for value in my_values]
 
# Change color
squarify.plot(sizes=my_values, alpha=.8, color=colors )
plt.axis('off')
plt.show()
```
